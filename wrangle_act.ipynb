```python
# create tweet_json.txt file if it is no already existing
file = 'tweet-json.txt'
if not os.path.isfile(file):
    open(file, 'w', encoding = 'UTF-8')
```

```python
# Get a list of tweet_ids from the enhanced twitter archive to use for 
# downloading with the Twitter API

# get start time of query
print("Start time:", datetime.datetime.now().time())
start = time.time()

# write JSON to .txt file
with open('tweet_json.txt', 'w', encoding = 'UTF-8') as file:
    for tweet_id in tweet_ids:
        try:
            tweet = api.get_status(tweet_id, tweet_mode = 'extended')
            json.dump(tweet._json, file)
            file.write('\n')
        except Exception as e:
            print("Error in Tweet ID:", tweet_id, "Time:",  datetime.datetime.now().time())

# get end time of query
end = time.time()
print("End time:", datetime.datetime.now().time())

# display runtime
print("Runtime: ", end - start)
```

```python
# read in the JSON data from the text file, and save to a DataFrame
import json
tweet_json_data = []

with open('tweet_json', 'r') as json_file:
    # read the first line to start the loop
    line = json_file.readline()
    while line:
        data = json.loads(line)

        # extract variables from the JSON data
        data_id = data['id']
        data_retweet_count = data['retweet_count']
        data_favorite_count = data['favorite_count']
        
        # create a dictionary with the JSON data, then add to a list
        json_data = {'tweet_id': data_id, 
                     'retweet_count': data_retweet_count, 
                     'favorite_count': data_favorite_count
                    }
        tweet_json_data.append(json_data)

        # read the next line of JSON data
        line = json_file.readline()
        # ----- while -----

# convert the tweet JSON data dictionary list to a DataFrame
twitter_api_data = pd.DataFrame(tweet_json_data, 
                                   columns = ['tweet_id',
                                              'retweet_count',
                                              'favorite_count'])
```

## Define 
* Create gender column

### Code

```python
# all kinds of gender expression in 'text' column
male = ['He', 'he', 'Him', 'him', "He's", "he's", 'His', 'his', 'Himself', 'himself', ]
female = ['She', 'she', 'Her', 'her', 'Hers', 'hers', 'Herself', 'herself', "She's", "she's"]

gender = []

# design a function to extract gender info
def gender_fun(data):
    for text in data.text:
        if (set(text.split()) & set(male)): 
            gender.append('male')
        elif (set(text.split()) & set(female)):
            gender.append('female')
        else:
            gender.append('None')
            

gender_fun(full_twitter_data)

# add gender column and attribute the values from gender list
full_twitter_data['gender'] = gender

# Convert 'None' to NA
full_twitter_data.gender = full_twitter_data.gender.replace('None', np.nan)
```

## Define
 * Simplify the table by keeping only one prediction, according to the odds priority order is as p1 > p2 > p3.

## Code

```python
predictions = []
confidence = []

# store the fisrt true algorithm with it's odds

# dog_prediction_confidence function:
# find the first true algorithm and append it to a list with it's odds
# if flase, predictions list will have values of NaN
def dog_prediction(data):
        if data.p1_dog == True:
            predictions.append(data.p1)
            confidence.append(data.p1_conf)
        elif data.p2_dog == True:
            predictions.append(data.p2)
            confidence.append(data.p2_conf)
        elif data.p3_dog == True:
            predictions.append(data.p3)
            confidence.append(data.p3_conf)
        else: 
            predictions.append(np.nan)
            confidence.append(0)
            
full_twitter_data.apply(dog_prediction, axis = 1)
full_twitter_data['predictions'] = predictions
full_twitter_data['confidence'] = confidence
```
